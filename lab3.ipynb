## Метод Крамера

Найти решение СЛАУ при помощи метода Крамера.
* x-2y+3z-t=6
* 2x+3y-4z+4t=-7
* 3x+y-2z-2t=9
* x-3y+7z+6t=-7


Вычисляем определитель матрицы системы:

```python
K = np.array([[1., -2., 3., -1.], [2., 3., -4., 4.], [3., 1., -2., -2.], [1., -3., 7., 6.]])
L = np.array([[6.], [-7.], [9.], [-7.]])
```

```python
D = round(np.linalg.det(K), 3)
D
```

```python
J = np.copy(K)
J[:, 0] = L[:, 0]
J
```

```python
D_1 = np.linalg.det(J)
D_1
```

```python
J = np.copy(K)
J[:, 1] = L[:, 0]
```

```python
D_2 = np.linalg.det(J)
D_2
```

```python
J = np.copy(K)
J[:, 2] = L[:, 0]
```

```python
D_3 = np.linalg.det(J)
D_3
```

```python
J = np.copy(K)
J[:, 3] = L[:, 0]
```

```python
D_4 = np.linalg.det(J)
D_4
```
